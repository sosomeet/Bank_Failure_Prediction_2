```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ---------------------------------
# 1) 데이터 로드
# ---------------------------------
df = pd.read_csv("training_data.csv")

# ---------------------------------
# 2) asset_turnover 추출 (원본)
# ---------------------------------
x = pd.to_numeric(df["asset_turnover"], errors="coerce").dropna()

# ---------------------------------
# 3) 기본 통계량
# ---------------------------------
mu = x.mean()
sigma = x.std(ddof=1)

print("[asset_turnover raw distribution]")
print("count :", len(x))
print("mean  :", mu)
print("std   :", sigma)
print("min   :", x.min())
print("max   :", x.max())
print("p25   :", np.percentile(x, 25))
print("p50   :", np.percentile(x, 50))
print("p75   :", np.percentile(x, 75))

# ---------------------------------
# 4) 분포 그래프
# ---------------------------------
plt.figure(figsize=(8, 6))
plt.hist(x, bins=40, density=True, alpha=0.7)
plt.axvline(mu, linestyle="--", label=f"mean = {mu:.3f}")
plt.axvline(mu + sigma, linestyle=":", label="+1 std")
plt.axvline(mu - sigma, linestyle=":", label="-1 std")

plt.xlabel("asset_turnover (raw)")
plt.ylabel("Density")
plt.title("Raw distribution of asset_turnover")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
